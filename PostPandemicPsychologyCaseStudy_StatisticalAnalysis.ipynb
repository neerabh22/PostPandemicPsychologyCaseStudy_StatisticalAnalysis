```python
import pandas as pd  # data processing, CSV file I/O
import seaborn as sns #visualisation
survey = pd.read_csv('Post PandemicPsychologyCaseStudy.csv')
survey.head(20)
```

```python
notes = survey[['Timestamp', 'Unnamed: 16','0']]
survey.drop(notes, axis =1, inplace=True)
```

```python
survey.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
```

```python
flatUI = ["#2c2c54", "#34ace0", "#84817a", "#ff3f34", "#05c46b", "#ffa801"]
sns.set_palette(flatUI)
sns.palplot(sns.color_palette())
plt.show()

def showFigure(fig, x=16, y=4):
    fig = plt.gcf()
    fig.set_size_inches(x, y)
    plt.show()
```

```python
survey.columns = ["Geade_Level", "Gender","MentalHEalth_ScaleRating_CurrentSchoolYear","HardestClass","WhyHardestClass","WishDifferentForRemortLearning","TeacherHelpful","HowSuccessRemortLerning","FamilySupport","FriendSupport","HoeToBetterMentalHealth","LevelOfAnxious","DailyScreenTime","DailySleepTime","InterestingAboutYourself"]
survey.head(10)
```

```python
survey['DailySleepTime'] = survey['DailySleepTime'].str.replace(r'\D', '')
survey['DailyScreenTime'] = survey['DailyScreenTime'].str.replace(r'\D', '')
survey.head(5)
```

```python
## Counting gender classes
survey.Gender.value_counts()
```

```python
## Counting Geade_Level classes
survey.Geade_Level.value_counts()
```

```python
survey.info()
```

```python
survey['DailySleepTime']=pd.to_numeric(survey['DailySleepTime'])
survey['DailyScreenTime']=pd.to_numeric(survey['DailyScreenTime'])

survey.info()
```

```python
fig = sns.distplot(survey['MentalHEalth_ScaleRating_CurrentSchoolYear']);    
showFigure(fig)    
fig = sns.countplot(x="Gender", data=survey)    
showFigure(fig)
fig = sns.countplot(x='Geade_Level', data=survey)    
showFigure(fig)

fig = sns.boxenplot(x='Geade_Level', y='MentalHEalth_ScaleRating_CurrentSchoolYear', data=survey)
showFigure(fig)
```

```python
fig = sns.catplot(x='Geade_Level', hue='Gender', data=survey, kind="count")
showFigure(fig)
fig = sns.catplot(x='Geade_Level', hue='FamilySupport', data=survey, kind="count")
showFigure(fig)
fig = sns.catplot(x='Geade_Level', hue='FriendSupport', data=survey, kind="count")
showFigure(fig)
fig = sns.catplot(x='Geade_Level', hue='HardestClass', data=survey, kind="count")
showFigure(fig)
fig = sns.catplot(x='Geade_Level', hue='MentalHEalth_ScaleRating_CurrentSchoolYear', data=survey, kind="count")
showFigure(fig)
#fig = sns.catplot(x='Geade_Level', hue='Rate Reaction to Problems', data=survey, kind="count")
#showFigure(fig)
fig = sns.swarmplot(x='Geade_Level', y='MentalHEalth_ScaleRating_CurrentSchoolYear', hue='Gender', palette="bright", data=survey)
showFigure(fig, y=6)
fig = sns.catplot(x='Geade_Level', y='FamilySupport', hue='FriendSupport', kind="bar", data=survey)
showFigure(fig)
```

```python
# Plot histogram grid
survey.hist(figsize=(16,16), xrot=-45) ## Display the labels rotated by 45 degress

# Clear the text "residue"
plt.show()
```
